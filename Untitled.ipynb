```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Yearly.csv',index_col='Year')
```

```python
df
```

```python
train = df.iloc[:-1,:]
test = df.iloc[-1,:]
```

```python
testnp = test.to_numpy()
```

```python
testnp
```

```python
train
```

```python
X_train = train.drop(columns=['HOMICIDE AND RELATED OFFENCES'])
y_train = train['HOMICIDE AND RELATED OFFENCES']
```

```python
X_test = np.delete(testnp, 2)
y_test = np.array([testnp[2]])
```

```python
print(X_test)
print(y_test)
```

```python
X_test=X_test.reshape(1, -1)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=5)
rf.fit(X_train.values, y_train.values)
y_pred = rf.predict(X_test)
f"Predicted{y_pred} Original{y_test}"
```

```python
from sklearn import svm
regr = svm.SVR()
regr.fit(X_train.values, y_train.values)
y_pred = regr.predict(X_test)
f"Predicted{y_pred} Original{y_test}"
```

```python
from sklearn.linear_model import SGDClassifier
sgdr = SGDClassifier(loss="squared_error", penalty="l2", max_iter=100)
sgdr.fit(X_train.values, y_train.values)
y_pred = sgdr.predict(X_test)
f"Predicted{y_pred} Original{y_test}"
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train.values, y_train.values)
y_pred = reg.predict(X_test)
f"Predicted{y_pred} Original{y_test}"
```
